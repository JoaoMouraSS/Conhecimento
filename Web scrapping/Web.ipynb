```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://informatica.mercadolivre.com.br/portateis-e-acessorios-notebooks/apple/usado/apple-macbook_NoIndex_True#applied_filter_id%3DITEM_CONDITION%26applied_filter_name%3DCondi%C3%A7%C3%A3o%26applied_filter_order%3D16%26applied_value_id%3D2230581%26applied_value_name%3DUsado%26applied_value_order%3D3%26applied_value_results%3D4261%26is_custom%3Dfalse'
headers = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.64"}

req = Request(url, headers= headers)
reposta = urlopen(req)

dados = reposta.read().decode('utf-8')

soup = BeautifulSoup(dados, 'html.parser')
```

```python
resultadoF = []

anuncio = soup.find('div', {'class': 'ui-search-main ui-search-main--only-products'}).findAll('div', {'ui-search-result__wrapper'})

for anuncio in anuncio:
    resultado = {}
    
    resultado['valor'] = (anuncio.find('span', {'class':'price-tag-fraction'}).getText())
    resultado['titulo'] = anuncio.find('h2', {'class': 'ui-search-item__title ui-search-item__group__element'})
    
    resultadoF.append(resultado)
```

```python
dataset = pd.DataFrame(resultadoF)
dataset
```
